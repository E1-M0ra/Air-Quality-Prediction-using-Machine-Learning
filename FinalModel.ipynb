```python
import pandas as pd
from lightgbm import LGBMClassifier
import joblib
```

```python
df = pd.read_csv(r"updated_pollution_dataset.csv")
```

```python
x = df.drop(columns='Air Quality')
y = df['Air Quality']
```

```python
model = LGBMClassifier()
model.fit(x, y)
```

```python
joblib.dump(model, "Air_Quality_Prediction.pkl")
```
